```python
def remove_pattern(text, pattern_regex):
    r = re.findall(pattern_regex, text)
    for i in r:
        text = re.sub(i, '', text)
    
    return text 
```

```python
#newline replaced by spaces

df_clean.tweet = df_clean.tweet.apply(lambda x: re.sub(r"\n", ' ', x))
# Count #HASHTAG and @MENTIONS:

#df_clean.tweet = df_clean.tweet.apply(lambda x : ([t for t in x.split() if t.startswith('#')]))
#df_clean.tweet = df_clean.tweet.apply(lambda x : re.sub(r"[for t in tweet: if t.startswith('@')]"," ",x))
#df_clean
df_clean['tweet'] = np.vectorize(remove_pattern)(df_clean['tweet'], "@[\w]*")
df_clean
```

```python
#english nos and text(all links, hashtags, mentions)
df_clean.tweet = df_clean.tweet.apply(lambda x: re.sub(r"[0-9]", ' ', x))
#marathi nos and fullstops etc
df_clean.tweet = df_clean.tweet.apply(lambda x: re.sub(r"[१२३४५६७८९०₹•…॥।]", " ", x))
```

```python
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"              "]+", re.UNICODE)
    return re.sub(emoj, '', data)

df_clean.tweet = df_clean.tweet.apply(remove_emojis)
```

```python
def remove_punctuations(tweet):
    for punctuation in string.punctuation:
        tweet = tweet.replace(punctuation, ' ')
    return tweet

df_clean.tweet = df_clean.tweet.apply(remove_punctuations)
```

```python
#multiple spaces -> single space
df_clean.tweet = df_clean.tweet.apply(lambda x: re.sub(r"  +", " ", x))
```
